```python
from huggingface_hub import login
from dotenv import load_dotenv
import os

load_dotenv()
login(token = os.getenv('your_huggingface_token'))
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "skt/ko-gpt-trinity-1.2B-v0.5"

model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# 입력 프롬프트 설정
prompt = "어린이를 위한 짧은 동화를 만들어주세요."

# 입력을 토큰화
inputs = tokenizer(prompt, return_tensors="pt")

# 모델을 사용하여 텍스트 생성
output = model.generate(**inputs, max_length=200, do_sample=True, top_k=50, top_p=0.95)

# 생성된 텍스트 디코딩
story = tokenizer.decode(output[0], skip_special_tokens=True)

print("생성된 동화:")
print(story)
```

```python
# 입력 프롬프트 설정
prompt = "질문: 선인장 키우는 방법을 설명해주세요.\n### 답변:"

# 입력을 토큰화
inputs = tokenizer(prompt, return_tensors="pt")

# 모델을 사용하여 텍스트 생성
output = model.generate(
    **inputs,
    max_length=200,
    do_sample=True,
    top_k=50,
    top_p=0.9,  # top_p 줄이기
    temperature=0.7,  # temperature 낮추기 (더 논리적인 답변 유도)
    repetition_penalty=1.2  # 반복되는 이상한 문장 방지
)

# 생성된 텍스트 디코딩
story = tokenizer.decode(output[0], skip_special_tokens=True)

print("생성된 답변:")
print(story)
```
