```python
import json
from groq import Groq

# Groq API 키 설정
GROQ_API_KEY = "your_api_key"  # 🔒 API 키를 여기에 입력하세요.

# 질의응답 자동 생성 (30개 생성)
def generate_qa(text):
    prompt = f"""
    SFT를 위한 학습 데이터를 생성하려고 합니다.
    다음 내용을 기반으로 30개의 한국어 Q&A 세트를 JSON 리스트 형태로 만들어 주세요.

    ⚠️ 반드시 JSON 형식만 출력하세요. 불필요한 설명을 포함하지 마세요.

    📌 형식 예시:
    [
        {{"question": "분갈이는 언제 해야 하나요?", "answer": "분갈이는 일반적으로 봄과 가을이 적기입니다."}},
        {{"question": "분갈이 후 주의할 점은?", "answer": "분갈이 후에는 충분한 물을 주고, 직사광선을 피하는 것이 중요합니다."}}
    ]

    입력 데이터:
    {text}
    """

    client = Groq(api_key=GROQ_API_KEY)

    # Groq 챗봇 API 호출
    response = client.chat.completions.create(
        model="mixtral-8x7b-32768",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7
    )

    # 응답 내용 추출
    raw_response = response.choices[0].message.content.strip()

    # JSON 유효성 검사 후 변환
    try:
        qa_list = json.loads(raw_response)
    except json.JSONDecodeError as e:
        print("⚠️ JSON 변환 오류:", e)
        print("📌 응답 내용:", raw_response[:500])  # 응답 앞부분 출력 (디버깅용)
        qa_list = []

    return qa_list

# HTML에서 텍스트 추출 후 GPT에 입력 (실제 text 사용)
qa_list = generate_qa("분갈이가 필요한 때와 분갈이 시기는?")

# JSON 저장 (올바른 리스트 형태)
with open("분갈이_2.json", "w", encoding="utf-8") as json_file:
    json.dump(qa_list, json_file, ensure_ascii=False, indent=4)

print(f"✅ Groq 기반 JSON 생성 완료! (Q&A {len(qa_list)}개 저장)")
```

```python
import os
import json

# 📂 JSON 파일이 있는 폴더 경로
folder_path = "."  # 실제 폴더 경로로 변경하세요

# 🔄 병합할 JSON 데이터를 저장할 리스트
merged_data = []

# 📂 "분갈이_"로 시작하는 JSON 파일만 가져오기
for filename in os.listdir(folder_path):
    if filename.lower().startswith("분갈이_") and filename.lower().endswith(".json"):
        file_path = os.path.join(folder_path, filename)

        # 🔍 JSON 파일 불러오기
        with open(file_path, "r", encoding="utf-8") as file:
            try:
                data = json.load(file)
                if isinstance(data, list):  # 리스트 형태인지 확인 후 병합
                    merged_data.extend(data)
                else:
                    print(f"⚠️ {filename} 파일은 리스트 형식이 아닙니다. 무시됨.")
            except json.JSONDecodeError as e:
                print(f"❌ JSON 디코딩 오류: {filename} - {e}")

# 📂 병합된 JSON 저장
output_file = "분갈이_merged.json"
with open(output_file, "w", encoding="utf-8") as outfile:
    json.dump(merged_data, outfile, ensure_ascii=False, indent=4)

print(f"✅ 병합 완료! 총 {len(merged_data)}개의 Q&A가 저장됨: {output_file}")
```

```python
import json
from collections import defaultdict

# 📂 JSON 파일 불러오기
file_path = "분갈이_merged.json"

try:
    with open(file_path, "r", encoding="utf-8") as file:
        qa_list = json.load(file)  # JSON 파일 로드

    # 📝 모든 질문(question)과 답변(answer) 수집
    questions_dict = defaultdict(list)

    for qa in qa_list:
        if "question" in qa and "answer" in qa:
            questions_dict[qa["question"]].append(qa["answer"])

    # 🔍 중복 질문 확인 및 출력
    duplicates = {question: answers for question, answers in questions_dict.items() if len(answers) > 1}

    if duplicates:
        print("🚨 중복된 질문 및 답변 목록:")
        for question, answers in duplicates.items():
            print(f"\n질문: {question}")
            for idx, answer in enumerate(answers, 1):
                print(f"  답변 {idx}: {answer}")
    else:
        print("✅ 중복된 질문이 없습니다.")

except FileNotFoundError:
    print(f"❌ 파일을 찾을 수 없습니다: {file_path}")
except json.JSONDecodeError as e:
    print(f"❌ JSON 디코딩 오류: {e}")
```

```python
import json

# 📂 JSON 파일 불러오기
file_path = "분갈이_merged.json"
output_file = "분갈이_merged_final.json"

try:
    with open(file_path, "r", encoding="utf-8") as file:
        qa_list = json.load(file)  # JSON 파일 로드

    # 📝 질문에 대해 나중에 나온 답변만 남기기
    questions_dict = {}

    for qa in qa_list:
        if "question" in qa and "answer" in qa:
            question = qa["question"]
            answer = qa["answer"]
            # 질문에 대한 마지막 답변만 업데이트
            questions_dict[question] = answer

    # 중복된 질문을 제외하고 새로운 리스트 생성
    unique_qa_list = [{"question": question, "answer": answer} for question, answer in questions_dict.items()]

    # 📂 최종적으로 수정된 JSON 저장
    with open(output_file, "w", encoding="utf-8") as outfile:
        json.dump(unique_qa_list, outfile, ensure_ascii=False, indent=4)

    print(f"✅ 중복 질문을 제외하고 마지막 답변만 저장 완료! 파일명: {output_file}")

except FileNotFoundError:
    print(f"❌ 파일을 찾을 수 없습니다: {file_path}")
except json.JSONDecodeError as e:
    print(f"❌ JSON 디코딩 오류: {e}")
```

```python
import json
from collections import defaultdict

# 📂 JSON 파일 불러오기
file_path = "홈가드닝_merged.json"

try:
    with open(file_path, "r", encoding="utf-8") as file:
        qa_list = json.load(file)  # JSON 파일 로드

    # 📝 모든 질문(question)과 답변(answer) 수집
    questions_dict = defaultdict(list)

    for qa in qa_list:
        if "question" in qa and "answer" in qa:
            questions_dict[qa["question"]].append(qa["answer"])

    # 🔍 중복 질문 확인 및 출력
    duplicates = {question: answers for question, answers in questions_dict.items() if len(answers) > 1}

    if duplicates:
        print("🚨 중복된 질문 및 답변 목록:")
        for question, answers in duplicates.items():
            print(f"\n질문: {question}")
            for idx, answer in enumerate(answers, 1):
                print(f"  답변 {idx}: {answer}")
    else:
        print("✅ 중복된 질문이 없습니다.")

except FileNotFoundError:
    print(f"❌ 파일을 찾을 수 없습니다: {file_path}")
except json.JSONDecodeError as e:
    print(f"❌ JSON 디코딩 오류: {e}")
```

```python
import os
import json

# 📂 병합할 JSON 파일 목록
files_to_merge = ["분갈이_merged.json", "홈가드닝_merged.json", "Q&A_merged.json",
                  "train_groq(물갈이).json", "train_groq(도시농업).json", "train_groq(건조형 실내정원의 실제).json", "train_groq(나만의 텃밭 가꾸기).json"]

# 🔄 병합할 JSON 데이터를 저장할 리스트
merged_data = []

# 📂 각 파일을 하나씩 불러와서 병합
for filename in files_to_merge:
    if os.path.exists(filename):
        print(f"📄 {filename} 파일을 병합 중...")

        # 🔍 JSON 파일 불러오기
        with open(filename, "r", encoding="utf-8") as file:
            try:
                data = json.load(file)

                # ✅ 데이터가 리스트 형태인지 확인 후 병합
                if isinstance(data, list):
                    merged_data.extend(data)
                else:
                    print(f"⚠️ {filename}의 데이터 구조가 예상과 다릅니다. 무시됨.")
            except json.JSONDecodeError as e:
                print(f"❌ JSON 디코딩 오류: {filename} - {e}")
    else:
        print(f"🚨 {filename} 파일이 존재하지 않습니다.")

# 📂 병합된 JSON 저장
output_file = "sft_dataset_1.json"
with open(output_file, "w", encoding="utf-8") as outfile:
    json.dump(merged_data, outfile, ensure_ascii=False, indent=4)

# 📊 총 데이터셋 개수 확인
print(f"✅ 병합 완료! 총 {len(merged_data)}개의 데이터셋이 저장됨: {output_file}")
```
